```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
```

```python
data0 = pd.read_csv("data/mine_explore0_inverted-pendulum_06-12-2023_13-32-33//data.csv")
data5 = pd.read_csv("data/mine_explore5_inverted-pendulum_06-12-2023_13-35-03//data.csv")
data10 = pd.read_csv("data/mine_explore10_inverted-pendulum_06-12-2023_13-37-39//data.csv")
data15 = pd.read_csv("data/mine_explore15_inverted-pendulum_06-12-2023_13-40-18//data.csv")
data20 = pd.read_csv("data/mine_explore20_inverted-pendulum_06-12-2023_13-43-11//data.csv")
datann = pd.read_csv("data/mine_explore15_inverted-pendulum_06-12-2023_13-46-17/data.csv")
```

```python
plt.plot(datann['itr'], datann['Train_AverageReturn'], label='NN')
plt.plot(data0['itr'], data0['Train_AverageReturn'], label='GP')
plt.plot(data5['itr'], data5['Train_AverageReturn'], label='GP + Explore 5')
plt.plot(data10['itr'], data10['Train_AverageReturn'], label='GP + Explore 10')
plt.plot(data15['itr'], data15['Train_AverageReturn'], label='GP + Explore 15')
plt.plot(data20['itr'], data20['Train_AverageReturn'], label='GP + Explore 20')
plt.grid(True)
plt.xlabel('Iteration')
plt.ylabel('Average Return')
plt.title('Inverted Pendulum')
plt.legend()
```

```python
plt.figure(figsize=(10,6))
plt.plot(datann['itr'], datann['Eval_AverageReturn'], label='NN')
plt.plot(data0['itr'], data0['Eval_AverageReturn'], label='GP')
plt.plot(data5['itr'], data5['Eval_AverageReturn'], label='GP + Explore 5')
plt.plot(data10['itr'], data10['Eval_AverageReturn'], label='GP + Explore 10')
plt.plot(data15['itr'], data15['Eval_AverageReturn'], label='GP + Explore 15')
plt.plot(data20['itr'], data20['Eval_AverageReturn'], label='GP + Explore 20')
plt.grid(True)
plt.xlabel('Iteration')
plt.ylabel('Average Return')
plt.title('Inverted Pendulum')
plt.legend()
plt.savefig("eval.png")
```
